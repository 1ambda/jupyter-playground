## 법정동 코드

- (법정동 코드) https://www.code.go.kr/index.do

파일 다운로드 후 CP949 txt 를 로딩해 아래 코드를 통해 UTF-8 CSV 로 저장하기

```python
## 법정동 코드 (CP949) 를 UTF-8 CSV 로 저장하기
def convertDistrictCodeB():
    p_ymd = '20200117'
    
    rawFilePath = './dataset/district_code_b_{}.txt'.format(p_ymd)
    csvFilePath = './dataset/district_code_b_{}.csv'.format(p_ymd)
    csvFilePathLatest = './dataset/district_code_b.csv'

    dfRaw = pd.read_csv(
        rawFilePath, 
        sep = '\t',
        encoding = 'CP949')
    
    # 컬럼 이름 변경
    dfRaw.rename(columns = {'법정동명': '법정동',}, inplace = True)
    
    # 컬럼 타입 변경
    dfRaw[['법정동코드']] = dfRaw[['법정동코드']].apply(pd.to_numeric, errors='coerce')

    # 컬럼 값 변경
    dfRaw = dfRaw.replace({ '폐지여부': {'존재': False, '폐지': True}})
    
    # p_ymd 추가
    dfRaw['p_ymd'] = p_ymd
    
    dfRaw.to_csv(csvFilePath, mode='w', index=False, encoding='utf8')
    dfRaw.to_csv(csvFilePathLatest, mode='w', index=False, encoding='utf8')

    return dfRaw
```

```python
dfDistrictCodeB = convertDistrictCodeB()
dfDistrictCodeB
```

## 행정동 - 법정동 매핑 코드

- (통계청) https://kssc.kostat.go.kr:8443/ksscNew_web/index.jsp#

파일 다운로드 후, 법정동 연계 시트를 CSV (파일 이름은 .txt) 저장 및 첫 라인 삭제

```python
def convertDistrictCodeH():
    
    p_ymd = '20200120'
    
    rawFilePath = './dataset/district_code_h_{}.txt'.format(p_ymd)
    csvFilePath = './dataset/district_code_h_{}.csv'.format(p_ymd)
    csvFilePathLatest = './dataset/district_code_h.csv'

    dfRaw = pd.read_csv(
        rawFilePath, 
        encoding = 'UTF-8')
    
    # p_ymd 추가
    dfRaw['p_ymd'] = p_ymd

    dfRaw.to_csv(csvFilePath, mode='w', index=False, encoding='utf8')
    dfRaw.to_csv(csvFilePathLatest, mode='w', index=False, encoding='utf8')
    
    return dfRaw
```

```python
dfDistrictCodeH = convertDistrictCodeH()
dfDistrictCodeH
```
